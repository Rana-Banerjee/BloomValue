```python
# Find low frequency codes
def calculate_max_words(df, threshold):
    total_unique_codes = len(np.unique(df.values))
    # Set the threshold
    #  Combine all columns into a single series and ignore empty string values
    if isinstance(df, pd.DataFrame):
        all_values = pd.concat([df[col] for col in df.columns])
    elif isinstance(df, pd.Series):
        all_values =df
    all_values = all_values[all_values != '']
    # Count the occurrences of each unique value
    value_counts = all_values.value_counts()
    # Filter codes that occur more than the threshold
    filtered_codes = value_counts[value_counts > threshold].index
    # Num filtered codes above threshold
    max_words = len(filtered_codes)
    print(f"Total unique codes: {total_unique_codes}")
    print(f"Number of codes occurring {threshold} times or less: {total_unique_codes- max_words}")
    print(f"max_words = Number of codes occurring more than {threshold} times: {max_words}")
    print(f"Coverage with threshold = {threshold} and max_words = {max_words} : {round(max_words/total_unique_codes * 100,2)} %")
    return max_words
```

```python
# Tokenize from words to integers (sequences) 
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

def build_tokenizer(texts, max_words):
    max_words = max_words # considers only the top number of words
    # max_features = max_words
    tokenizer = Tokenizer(num_words=max_words)
    tokenizer.fit_on_texts(texts)
    word_index = tokenizer.word_index
    print('Found %s unique tokens.' % len(word_index))
    return tokenizer

def tokenize_df(df, tokenizer):
    if isinstance(df, pd.DataFrame):
        for col in df.columns:
            tokenized_cols = tokenizer.texts_to_sequences(df[col].tolist())
            data = pad_sequences(tokenized_cols, maxlen=1, padding='post', truncating='post', value=0.0)
            data = pd.DataFrame(data)
            df[col]=data
    elif isinstance(df, pd.Series):
        tokenized_cols = tokenizer.texts_to_sequences(df.values)
        data = pad_sequences(tokenized_cols, maxlen=1, padding='post', truncating='post', value=0.0)
        data = pd.DataFrame(data)
        df=data
    return df
```
