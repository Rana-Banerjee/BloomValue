```python
# Convert columns to appropriate data types
date_columns=[
    'Service_Date',
    'Admission_Date',
    'Discharge_Hour'
]

int_columns =[
    'MIA_Covered_Days',
    'CoInsured_Days',
    'Non-Covered_Days',
    'Line_Serial',
    'Line_Units'
]

float_columns = [
    'Paid_Amount',
    'Charge_Amount',
    'MIA_Claim_PPS_Capital_Amount',
    'MIA_PPS_Capital_FSP_DRG_Amount',
    'MIA_Claim_PPS_Capital_Outlier_Amount',
    'MIA_Claim_Disproportionate_Share_Amount',
    'MIA_PPS_Capital_IME_Amount',
    'MIA_PPS_Operating_Outlier_Amount',
    'DRG_Weight',
    'Line_Amount_Total'
]
```

```python
for c in claims_data.columns:
    if c not in date_columns+int_columns+float_columns:
        claims_data[c] = claims_data[c].fillna('')
    else:
        print(c, claims_data[c].isna().sum())
```

```python
for c in claims_data.columns:
    if c in int_columns:
        claims_data[c]=claims_data[c].astype(int)
    elif c in float_columns:
        claims_data[c]=claims_data[c].astype(float)
    elif c not in date_columns:
        claims_data[c]=claims_data[c].astype(str)

for c in revenue_data.columns:
    if c in int_columns:
        revenue_data[c]=revenue_data[c].astype(int)
    elif c in float_columns:
        revenue_data[c]=revenue_data[c].astype(float)
    elif c not in date_columns:
        revenue_data[c]=revenue_data[c].astype(str)      
```

```python
# Format Date Types
claims_data['Service_Date'] = pd.to_datetime(claims_data['Service_Date'], format='%Y%m%d', errors='coerce')
claims_data['Admission_Date'] = pd.to_datetime(claims_data['Admission_Date'], format='%Y%m%d', errors='coerce')
claims_data['Discharge_Hour']= pd.to_datetime(claims_data['Discharge_Hour'].astype(int), format='%Y%m%d', errors='coerce')
revenue_data['Service_Date']= pd.to_datetime(revenue_data['Service_Date'], format='%d-%m-%Y', errors='coerce')
```

### Feature Extraction/Engineering

```python
# Revenue data feature engineering
# Add Length of stay
claims_data['LOS']=claims_data['Discharge_Hour']-claims_data['Admission_Date']
claims_data['LOS']= claims_data['LOS'].dt.days
claims_data['LOS'].head()
```

```python
# Add number of ICD and CPT codes
icd_cols = [c for c in claims_data.columns if 'ICD_DGNS' in c]
cpt_cols = [c for c in claims_data.columns if 'ICD_PRCDR' in c]

claims_data['ICD_Count'] = claims_data[icd_cols].apply(lambda row: row.ne('').sum(), axis=1)
claims_data['CPT_Count'] = claims_data[cpt_cols].apply(lambda row: row.ne('').sum(), axis=1)
```

```python
### Add more features as per domain inputs
```

```python
# Revenue data feature extraction
# Pivot data at revenue code level
# Frequency of a Revenue code across a claim
# Total_Units: Sum of Line_Units for each revenue code across a claim
# Total_Amount: Sum of Line_Amount_Total for each revenue code across all lines per claim
## Add other features as per domain inputs

agg_df = revenue_data.groupby([
    'Billed_Subscriber_ID',
    'Patient_Account_Number',
    'Service_Date',
    'Line_Revenue_Code'
], as_index=False).agg(
    {
        'Line_Serial' : 'size',
        'Line_Units': 'sum',
        'Line_Amount_Total': 'sum'
    }
).rename(columns={
    'Line_Serial': 'Total_Lines_Per_LRCode',
    'Line_Units': 'Total_Units_Per_LRCode',
    'Line_Amount_Total': 'Total_Amount_Per_LRCode'
}).reset_index()

# Pivot the data
pivot_df = agg_df.pivot(index=['Billed_Subscriber_ID','Patient_Account_Number','Service_Date'], columns='Line_Revenue_Code', values=['Total_Lines_Per_LRCode', 'Total_Units_Per_LRCode', 'Total_Amount_Per_LRCode'])
# Flatten the MultiIndex columns
pivot_df.columns = [f'{col[0]}-{col[1]}' for col in pivot_df.columns]
pivot_df.reset_index(inplace=True)
pivot_df.fillna(0, inplace=True)
```
